## Initialize

```python
from cmbagent import CMBAgent
import os
cmbagent = CMBAgent(agent_llm_configs = {'engineer': {
                        "model": "gpt-4.5-preview-2025-02-27",
                        "api_key": os.getenv("OPENAI_API_KEY"),
                        "api_type": "openai",
                        }})
```

## Solve

```python
task = r"""
Task: Uncovering Utility Functions with Observational Data

Objective:
Develop an analytical pipeline to estimate individual utility functions from observational data capturing consumer behavior. 
The solution should process simulated data representing consumers choosing among various goods under differing income levels and price conditions, 
mirroring market dynamics in an economy similar to the New York City Area.

Key steps include:
- Data preprocessing and cleaning
- Econometric modeling to estimate utility functions
- Visualization of estimated utility functions and consumer decision patterns

Do not implement API endpoints or use Flask; instead, provide standalone visualizations.
"""

cmbagent.solve(task,
               max_rounds=500,
               shared_context = {'feedback_left': 1,
                                 'maximum_number_of_steps_in_plan': 4})
```
