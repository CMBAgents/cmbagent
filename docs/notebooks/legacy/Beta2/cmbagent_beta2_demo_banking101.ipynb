```python
import os
from cmbagent import CMBAgent

cmbagent = CMBAgent(agent_llm_configs = {
                    'engineer': {
                        "model": "o3-mini-2025-01-31",
                        "reasoning_effort": "high", ## for gpt-4.5-preview-2025-02-27, gpt-4o-2024-11-20, gpt-4o-mini, etc, comment out this line.
                        "api_key": os.getenv("OPENAI_API_KEY"),
                        "api_type": "openai",
                        },
                    'researcher': {
                        "model": "gemini-2.0-pro-exp-02-05",
                        "api_key": os.getenv("GEMINI_API_KEY"),
                        "api_type": "google",
                        }})

task = """
Generate simulated customer behavior data for a bank's private equity division that caters to high-income individuals in the UK (e.g., Weatherbys, Coutts, Child & Co, C. Hoare & Co). 
The simulation should encompass around 10,000 customers over a 3-year period, incorporating realistic dynamics such as seasonal spending patterns, varying investment preferences, and external economic influences (e.g., sudden interest rate hikes, market downturns, or regulatory changes) that impact customer behavior.

Your simulation should capture:
-Diverse customer profiles based on demographics, net worth, risk tolerance, and investment horizon.
-Detailed behavioral patterns including account transactions, investment choices, frequency of high-value transactions, and responses to targeted bank offers.
-The influence of economic shocks on customer behavior, such as shifts in liquidity preferences, changes in asset allocation, and alterations in engagement with private equity products.
-Correlations among customer segments, highlighting dependencies between socio-economic factors and investment behavior.

After generating the simulated data, apply a behavioral analytics model to evaluate:
- Customer lifetime value (CLV) and segmentation strategies.
- The impact of economic disruptions on customer retention, investment performance, and overall engagement.


Provide visualizations that illustrate:
- The evolution of key customer metrics (e.g., account balances, transaction volumes) over time, with emphasis on periods of economic stress.
- Clustering results that highlight distinct customer segments and their behavioral patterns.
- Correlation structures among demographic attributes, investment choices, and transaction behaviors.
- Comparative analysis of customer behavior and portfolio performance during normal periods versus economic shocks.


Finally, analyze the simulation outcomes to discuss:
- How external economic conditions affect customer behavior in the high-income, private equity segment.
- Insights on risk management, customer segmentation, and investment strategies.
- Any limitations or biases in the simulation approach and the performance of the model(s) under extreme conditions. 
"""

cmbagent.solve(task,
               max_rounds=500, # set to a high number, this is the max number of total agent calls
               shared_context = {'feedback_left': 1, # number of feedbacks on the plan, generally want to set to a low number, as this adds unnecessary complexity to the workflow. 
                                 'maximum_number_of_steps_in_plan': 5})
```
