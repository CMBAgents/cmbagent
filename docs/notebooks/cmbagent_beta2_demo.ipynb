## Initialize

```python
from cmbagent import CMBAgent

cmbagent = CMBAgent()
```

## Solve

```python
task = r"""
Write a numerical solver for the harmonic oscillator. 
In a different step use the solver to generate 500 trajectories and save them.
Train a neural net for emulating the x(w,t) trajectories with 0<t<1 and frequency w that should be in an interesting range. 
"""
cmbagent.solve(task,
               max_rounds=100,
               # initial_agent = "control",
               shared_context = {'feedback_left': 1,
                                 'maximum_number_of_steps_in_plan': 5})
```
