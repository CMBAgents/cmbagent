## Initialize

```python
from cmbagent import CMBAgent

cmbagent = CMBAgent()
```

## Solve

```python
task = r"""
Task: Energy Cost Extraction Based on Geographic Coordinates

Objective:
Develop an efficient pipeline that determines the energy cost for a given location 
by matching user-supplied geographic coordinates with the corresponding energy operators service area. 

The solution should use generated simulated data representing energy operators and users in a region similar to England.

Do not implement API endpoints or use Flask; instead, provide standalone visualizations.
"""
cmbagent.solve(task,
               max_rounds=500,
               shared_context = {'feedback_left': 1,
                                 'maximum_number_of_steps_in_plan': 5})
```
