# Can we do some physics with cmbagent?
### In this notebook we will illustrate how LLM may be possible to solve physics problems in cosmology, focusing on ```classy_sz_agent```.
We will investigate three simple tasks:
1. Fitting the shape of the matter power spectrum.
2. How the shape of tSZ power spectrum changes with mass cuts and pressure profiles.
3. How the tSZ x galaxy power spectrum changes with HOD parameters.

## 1. A simple example: matter power spectrum $P(k)$

We all know that matter power spectrum scales as:

$$
P(k) \propto 
\begin{cases} 
k & k < k_\text{eq} \\
k^{-3} & k > k_\text{eq}
\end{cases}
$$
where $k_{eq}\approx 0.01 Mpc^{-1}$, $z_{eq} \approx 3300$.

Can cmbagent discover this physics?

````python
from cmbagent import CMBAgent
cmbagent = CMBAgent(agent_list = ['classy_sz'],make_vector_stores=['classy_sz'], verbose=True)
````

````python
subtask1 = """Compute the linear matter power spectrum P(k) as a function of k, using reasonable cosmological parameters"""
subtask2 = """
Locate the peak of the matter power spectrum that divides the matter power spectrum into two parts. 
What is the value of k in unit of Mpc^-1 at this peak?
"""
subtask3 = """
For small k (k much smaller than 10^{-3}) and large k (k much larger than 1) limits, fit a line in the log space for these part of the power spectrum.
What are the gradients of the best-fit lines? Can you explain the physical reasons why the powers of polynomials look like that?
"""
subtask4 = """
Plot this matter power spectrum, and display the fitted curve and the power of polynomials on the plot.
Only save one single plot, with the plot name 'matter_power_spectrum.png'
"""
subtask5 = """
Please retrieve as much information as possible in classy_sz_agent, this is very important.
"""
````

````python
cmbagent.clear_cache()
task = subtask1 + subtask2 + subtask3 + subtask4 + subtask5
cmbagent.solve(task)
````

````python
cmbagent.show_plot("matter_power_spectrum.png")
````

### We can see that cmbagent:
- correctly predicts the peak at $k=0.0106$ Mpc$^{-1}$.
- The gradient fits at small $k$ and large $k$ are 0.78 and -2.54 respectively, consistent with the theory (should be 1 and -3).

However, it costs much for the engineer to code this task. The output is also uncertain depending on how you phrase the prompt, 
sometimes need to try several times to get the most optimum prompt.
