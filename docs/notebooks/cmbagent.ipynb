## Computing with camb and planck

```python
from cmbagent import CMBAgent
cmbagent = CMBAgent(agent_list = ['camb','planck'],
                    make_vector_stores=['planck'], 
                    verbose=True)
                   
```

## Computing a likelihood value with cobaya, act, planck and classy_sz 

### Initialize

```python
from cmbagent import CMBAgent



agent_temperature = {
'classy_sz_agent': 0.00001,

'cobaya_agent': 0.000001,
    
'planck_agent': 0.000001,
    
'act_agent': 0.000001,
}

cmbagent = CMBAgent(agent_list = ['classy_sz', 'planck', 'act', 'cobaya'], 
                    verbose=True,
                    agent_temperature = agent_temperature,
                    # cache_seed=44
                   ) 
```

```python
# clear cach to restart from scratch
cmbagent.clear_cache()
```

### solve task

```python
task = """
Use cobaya with classy_sz to evaluate the ACT DR6 lensing likelihood for sigma8=0.8 and Omega_m=0.31. 
Other parameters will be set to baseline best-fit Planck 2018 values in LCDM.
"""

cmbagent.solve(task)
```

```python
cmbagent.restore()
```

## Reproducing ACT DR6 lensing analysis

### Run the mcmc

```python
from cmbagent import CMBAgent

cmbagent = CMBAgent(agent_list = ['classy_sz','act', 'cobaya'], 
                    verbose=True,
                    make_vector_stores = ['classy_sz','act', 'cobaya']
                   ) 
```

```python
cmbagent.clear_cache()

task = """
Derive cosmological parameter constraints from ACT DR6 CMB lensing data
"""

cmbagent.solve(task)
```

### Download original chains

```python

cmbagent = CMBAgent(agent_list = ['act'], 
                    verbose=True,
                   ) 

cmbagent.clear_cache()

task = """
Download the baseline ACT DR6 lensing chains and store them in a new folder called dr6_lensing_actoriginal 
"""

cmbagent.solve(task)
```

## Solving non-cosmology tasks with cmbagent

You can use cmbagent to solve non-cosmology tasks as well. 
In this case, set the agent_list empty. 
The system will then work with engineer, executor and planner. 


```python
from cmbagent import CMBAgent

cmbagent = CMBAgent(agent_list = [], 
                    verbose=True,
                   ) 
```

```python
cmbagent.solve("""
generate a noisy realization of 50 points around the linear relation b+w*x with b=0.2 and w=0.6, with x between 0 and 1.
""")
```

```python
cmbagent.show_plot('noisy_data.png')
```
