# Idea Generation


A [Planning and Control](https://arxiv.org/abs/2507.07257) strategy to generate research ideas with an **idea maker** and an **idea hater** agent. 


## Planning and Control


This is a common strategy that is at the basis of most *Deep Research/Coding* tools, like [Claude Code](https://www.anthropic.com/claude-code) or [ChatGPT Deep research](https://openai.com/index/introducing-deep-research/).  

The idea is simple and illustrated below. 

<img src="planning_and_control.png" alt="Planning and Control" width="400"/>


### Planning

The **main task** is broken down into sub-tasks by **planner agents**. The sequence of sub-tasks forms a **Plan**.

### Control

The Plan is executed step-by-step. 

## Example: Generate Research Ideas with Bank Customer Data

```python
import os
os.environ["CMBAGENT_DISABLE_DISPLAY"] = "false"
import cmbagent
root = "~/Desktop/"
work_dir = root+"cmbagent_dir"
```

```python
planner_append_instructions = r"""
Given these datasets, and information, make a plan according to the following instructions: 

- Ask idea_maker to generate 5 new research project ideas related to the datasets.
- Ask idea_hater to critique these ideas.
- Ask idea_maker to select and improve 2 out of the 5 research project ideas given the output of the idea_hater.
- Ask idea_hater to critique the 2 improved ideas. 
- Ask idea_maker to select the best idea out of the 2. 
- Ask idea_maker to report the best idea in the form of a scientific paper title with a 5-sentence description. 

The goal of this task is to generate a research project idea based on the data of interest. 
Don't suggest to perform any calculations or analyses here. The only goal of this task is to obtain the best possible project idea.
"""


idea_maker_model = "gpt-4.1-2025-04-14"
idea_hater_model = "claude-sonnet-4-20250514"


data_description = "Bank customer data during covid-19"

results = cmbagent.planning_and_control_context_carryover(data_description,
                      n_plan_reviews = 1,
                      max_plan_steps = 6,
                      idea_maker_model = idea_maker_model,
                      idea_hater_model = idea_hater_model,
                      plan_instructions=planner_append_instructions,
                      work_dir = work_dir,
                      clear_work_dir = True
                     )
```

## Post-Processing Traces

```python
import os
from IPython.display import display, Markdown

def list_dir_tree(startpath, max_depth=5):
    output = ""
    startpath = os.path.expanduser(startpath)
    for root, dirs, files in os.walk(startpath):
        # Limit depth
        depth = root[len(startpath):].count(os.sep)
        if depth > max_depth:
            continue
        indent = "    " * depth
        output += f"{indent}- `{os.path.basename(root)}/`\n"
        for f in files:
            output += f"{indent}    - {f}\n"
    return output

tree_md = list_dir_tree(work_dir)
display(Markdown(f"**Tree view of** `{work_dir}`\n\n{tree_md}"))
```
